```python
def image_loader(path:str)->np.array:
    return np.array(Image.open(path))
```

```python
def histogram_visualization(image_data,title_name):
  red_data,blue_data,green_data=image_data[:,:,0].flatten(),image_data[:,:,1].flatten(),image_data[:,:,2].flatten()
  fig,ax=plt.subplots(1,3,figsize=(12,4))
  data_temp=[red_data,blue_data,green_data]
  colors=["red","blue","green"]
  titles=['Red Channel', 'Blue Channel', "Green Channel"]
  for i in range(3):
    ax[i].hist(data_temp[i],bins=256,density=False,color=colors[i],alpha=0.7)
    ax[i].set_title(titles[i])
    ax[i].set_xlabel('Pixel Value')
    ax[i].set_ylabel('Frequency')
  plt.suptitle(title_name)
  plt.tight_layout()
  plt.show()
```

```python
def compare_x_channel(images:np.array,channel_name:str,image_name:str):
  channel_name=channel_name.lower()
  data_temp=[image for image in images]
  channel_value={
      "red":0,
      "blue":1,
      "green":2
  }
  channel_index=channel_value[channel_name]
  if len(images)>1:
    fig,ax=plt.subplots(1,len(images),figsize=(16,4))
    for index,data_per_channel in enumerate(data_temp):
      temp_pixel_value=data_per_channel[:,:,channel_index].flatten()
      ax[index].hist(temp_pixel_value,bins=256,density=False,color=channel_name,alpha=0.7)
      ax[index].set_title(image_name[index])
      ax[index].set_xlabel("Pixel Value")
      ax[index].set_ylabel("Frequency")
    plt.tight_layout()
    plt.show()
  else:
    raise ValueError("Number of images should be greater than 1")
```

```python
def Visualization_compare(paths,chn_name):
  data=[]
  temp=[]
  for path in paths:
    data.append(image_loader(path))
    temp.append(os.path.basename(path[:-4]))
  compare_x_channel(data,chn_name,temp)
```

```python
Visualization_compare(file_paths[:2],"green")
```

```python
def all_channel_compare(image:list,image_name:list):
  colors=["red","blue","green"]
  fig,ax=plt.subplots(1,3,figsize=(18,7))
  for index,images in enumerate(image):
    for i in range(3):
      histr = calcHist([images], [i], None, [256],[0,256],accumulate=True)
      ax[index].plot(histr, color=colors[i])
    ax[index].set_xlim([0, 256])
    ax[index].legend(colors)
    ax[index].set_title(f"{image_name[index]}")
    ax[index].set_xlabel("Pixel Value")
    ax[index].set_ylabel("Frequency")
  plt.tight_layout()
  plt.show()
```

```python
def Visualization_compare_all(paths):
  data=[]
  temp=[]
  for path in paths:
    data.append(cv2.imread(path))
    temp.append(os.path.basename(path[:-4]))
  all_channel_compare(data,temp)
```

```python
Visualization_compare_all(file_paths)
```
